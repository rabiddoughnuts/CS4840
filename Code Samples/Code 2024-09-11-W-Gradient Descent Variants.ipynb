# Limitation of traditional gradient descent
- Running slowly when training set is large
- High memory consuming

## 1-dimensional feature vector and 100 training samples 

```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100, 1) #X are 100 inputs with 1-dimensional feature randomly distributed in [0, 2)
y = 3 * X + np.random.uniform(0, 2, (100, 1)) #y are true label values

plt.plot(X, y, 'b.')
plt.xlabel("X", fontsize=18)
plt.ylabel("y", fontsize=14, rotation=0)
plt.axis([0, 2, 0, 10])
plt.show()
```

```python
import time

np.random.rand(100, 1) #y are true label values

eta = 0.1 #Learning rate
n_iter = 100 #Number of iterations for weight updates
d_train = 100 #Number of training samples

w = np.random.rand(1) #Weight vector

t1 = time.time()
for i in range(n_iter):
    gradient = 2/d_train * (w.dot(X.T) - y.T).dot(X)
    w = w - eta * gradient
    train_loss = 1/d_train * (w.dot(X.T) - y.T).dot((w.dot(X.T) - y.T).T)
    print('iteration: {}: weight = {}, train loss = {}'.format(i + 1, w, train_loss))
t2 = time.time()
print('the time used for optimization: {}'.format(t2 - t1))
```

# Stochastic gradient descent

```python
X = 2 * np.random.rand(10000, 1) #X are 10,000 inputs with 1-dimensional feature randomly distributed in [0, 2)
y = 3 * X + np.random.uniform(0, 2, (10000, 1)) #y are true label values

eta = 0.1 #Learning rate
n_epochs = 10 #Number of epochs for weight updates
d_train = 10000 #Number of training samples

w = np.random.rand(1) #Weight vector

t1 = time.time()
for epoch in range(n_epochs):
    for i in range(d_train):
        xi = X[i]
        yi = y[i]
        gradient = 2 * (w.dot(xi.T) - yi.T).dot(xi)
        w = w - eta * gradient

    train_loss = 1/d_train * (w.dot(X.T) - y.T).dot((w.dot(X.T) - y.T).T)
    print('epoch: {}: weight = {}, train loss = {}'.format(epoch + 1, w, train_loss))
    indices = np.random.permutation(len(X))
    X = X[indices]  #Shuffle the input X
    y = y[indices]  #Shuffle the input y

t2 = time.time()
print('the time used for optimization: {}'.format(t2 - t1))
```

```python
y_predict = X.dot(w.T)
plt.plot(X, y, "b.")
plt.plot(X, y_predict, "r-")
plt.xlabel("X", fontsize=18)
plt.ylabel("y", fontsize=14, rotation=0)
plt.axis([0, 2, 0, 10])
plt.show()
```

# Mini-batch gradient descent

```python
X = 2 * np.random.rand(10000, 1) #X are 10,000 inputs with 1-dimensional feature randomly distributed in [0, 2)
y = 3 * X + np.random.uniform(0, 2, (10000, 1)) #y are true label values

eta = 0.1 #Learning rate
n_epochs = 10 #Number of epochs for weight updates
d_train = 10000 #Number of training samples
n_batches = 64 #Batch size

w = np.random.rand(1) #Weight vector

t1 = time.time()
for epoch in range(n_epochs):
    for i in range(0, d_train, n_batches):
        xi = X[i:i + n_batches]
        yi = y[i:i + n_batches]
        gradient = 2 / n_batches * (w.dot(xi.T) - yi.T).dot(xi)
        w = w - eta * gradient
    
    train_loss = 1 / n_batches * (w.dot(xi.T) - yi.T).dot((w.dot(xi.T) - yi.T).T)
    print('epoch: {}: weight = {}, train loss = {}'.format(epoch + 1, w, train_loss))
    indices = np.random.permutation(len(X))
    X = X[indices]  #Shuffle the input X
    y = y[indices]  #Shuffle the input y

t2 = time.time()
print('the time used for optimization: {}'.format(t2 - t1))
```

```python
y_predict = X.dot(w.T)
plt.plot(X, y, "b.")
plt.plot(X, y_predict, "r-")
plt.xlabel("X", fontsize=18)
plt.ylabel("y", fontsize=14, rotation=0)
plt.axis([0, 2, 0, 10])
plt.show()
```
